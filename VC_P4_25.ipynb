```python
from ultralytics import YOLO

# Cargar el modelo entrenado
model = YOLO('runs/detect/train/weights/best.pt')

# Ejecutar validación
metrics = model.val()

# Mostrar métricas principales
print(f"mAP50-95: {metrics.box.map:.4f}")
print(f"mAP50: {metrics.box.map50:.4f}")
print(f"mAP75: {metrics.box.map75:.4f}")

# Precision, Recall y F1 por clase
print(f"\nPrecision (media): {metrics.box.mp:.4f}")
print(f"Recall (media): {metrics.box.mr:.4f}")

# F1 Score (calculado manualmente)
precision = metrics.box.mp
recall = metrics.box.mr
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
print(f"F1 Score: {f1_score:.4f}")

# Métricas por cada clase
print("\nMétricas por clase:")
for i, map_value in enumerate(metrics.box.maps):
    print(f"Clase {i}: mAP50-95 = {map_value:.4f}")
```
